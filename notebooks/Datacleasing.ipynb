## Limpieza de datos con Pandas

1- importar las librerías necesarias para el procesamiento de los datos.

```python
import numpy as np
import pandas as pd
import os
```

2- Lectura de los datos

```python
data=pd.read_csv('https://www.datos.gov.co/resource/k2b5-jms2.csv',encoding='utf-8')
```

```python
data
```

3- Agregar una columna

```python
data['tipo monumento']='NaN' # para este caso, se agrega una columna, ='NaN' cuando sea letras, se agrega aspostrofe, si son numeros sin apostrofe, si no se le pone es aparecerá "0"
data
```

4- Agregar multiples columnas con muchos registros

```python
# Crea una lista de 100 elementos con las categorías que quieres agregar
#En este caso, se agrega columnas con sus datos por país
nueva_columna = ['Colombia'] * 101 # Ejemplo de lista con categorías repetidas

# Agrega la nueva columna al DataFrame
data['pais'] = nueva_columna



#agregar columnas con sus datos por departamento
depto = ['Valle del Cauca']*101
data['departamento'] = depto



#agregar columnas con sus datos por municipio
pueblo = ['Palmira']*101
data['municipio'] = pueblo

data
```

5- Eliminar una columna

```python
#data.drop('decreto_resoluci_n',axis=1, inplace=True)
#data
```

6- Eliminar multiples columnas

```python
data.drop(['item','ubicaci_n'],axis=1, inplace=True)# es de anotar que se deben encerrar en corchetes, aplicando el tipo de dato lista en python, por que está encerrado en parentesis ()
data
```

7- Eliminar una fila

```python
data.drop(index=100,axis=1, inplace=True)#para eliminar  una fila
data
```

8- Eliminar multiples filas

```python
data.drop(index=[97,98,99],axis=1, inplace=True)#para eliminar multiples filas, simepre encerrar en corchetes.
data
```

9- Eliminar duplicados

```python
data.drop_duplicates()
```

10 -Rellenar valores faltantes

```python
# Rellenar los valores faltantes en la columna 'A' con la cadena 'sin información'
data['decreto_resoluci_n'] = data['decreto_resoluci_n'].fillna('por decreto')

data
```

11 -Eliminar una sola columna

```python
data.drop('tipo monumento',axis=1, inplace=True)
data
```

12 -Renombrar columna

```python
data.rename(columns={'decreto_resoluci_n': 'Normatividad',},inplace=True)
data
```
